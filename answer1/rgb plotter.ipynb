```python
def compute_histogram(image):
	
	image_height = image.shape[0]
	image_width = image.shape[1]
	image_channels = image.shape[2]
	print(f"{image_channels} , {image_height}, {image_width}")
	histogram = np.zeros([256, image_channels], np.int16)
	
	for x in range(0, image_height):
		for y in range(0, image_width):
			for c in range(0, image_channels):
					histogram[image[x,y,c], c] +=1
	
	return histogram

def plot_histogram(histogram):
	plt.figure()
	plt.title("Color Image Histogram")
	plt.xlabel("Intensity Level")
	plt.ylabel("Intensity Frequency")
	plt.xlim([0, 256])
	plt.plot(histogram[:,0],'r') # This is to Plot Blue Channel with Blue Color
	plt.plot(histogram[:,1],'g') # This is to Plot Green Channel with Green Color
	plt.plot(histogram[:,2],'b') # This is to Plot Red Channel with Red Color
	plt.savefig("Color_Histogram.jpg")
	plt.show();
    
```

```python
	histogram = compute_histogram(img)

	plot_histogram(histogram)
```

```python
im = Image.open("../Images/3/Lena.bmp")
w, h = im.size  
colors = im.getcolors(w*h)

def hexencode(rgb):
    r=rgb[0]
    g=rgb[1]
    b=rgb[2]
    return '#%02x%02x%02x' % (r,g,b)

for idx, c in enumerate(colors):
    plt.bar(idx, c[0], color=hexencode(c[1]))

plt.show()
```
